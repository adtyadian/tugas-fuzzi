```python
import numpy as np
import matplotlib.pyplot as plt

def triangle(x, a, b, c, d, e):

    if x < a:
        return d
    elif a <= x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    elif x > c:
        return e
    else:
        return 0


# range temperatur
t = np.linspace(10, 40, 400)

dingin = []
sedang = []
panas = []

for x in t:
    dingin.append(triangle(x, 15, 15, 25, 1, 0))
    sedang.append(triangle(x, 15, 25, 35, 0, 0))
    panas.append(triangle(x, 25, 35, 35, 0, 1))


# temperatur yang diuji
temp = 32.5

mu_dingin = triangle(temp, 15, 15, 25, 1, 0)
mu_sedang = triangle(temp, 15, 25, 35, 0, 0)
mu_panas = triangle(temp, 25, 35, 35, 0, 1)

print("Temperatur =",temp,"°C")
print("μ Dingin =",mu_dingin)
print("μ Sedang =",mu_sedang)
print("μ Panas =",mu_panas)

fig, ax = plt.subplots(3,1, figsize=(6,9))

# ---------------- DINGIN ----------------
ax[0].plot(t, dingin, color='red', linewidth=3)
ax[0].axvline(temp, linestyle='--', color='orange')
ax[0].text(22,1.02,"Dingin",color='red',fontsize=14)
ax[0].set_ylabel("μD(t)")
ax[0].set_xlim(10,40)
ax[0].set_ylim(0,1.1)
ax[0].set_xticks([20,30])
ax[0].grid(False)

# ---------------- SEDANG ----------------
ax[1].plot(t, sedang, color='green', linewidth=3)
ax[1].axvline(temp, linestyle='--', color='orange')
ax[1].text(24,1.02,"Sedang",color='green',fontsize=14)
ax[1].set_ylabel("μS(t)")
ax[1].set_xlim(10,40)
ax[1].set_ylim(0,1.1)
ax[1].set_xticks([20,30])
ax[1].grid(False)

# ---------------- PANAS ----------------
ax[2].plot(t, panas, color='blue', linewidth=3)
ax[2].axvline(temp, linestyle='--', color='orange')
ax[2].text(33,1.02,"Panas",color='blue',fontsize=14)
ax[2].set_ylabel("μP(t)")
ax[2].set_xlabel("Temperatur t")
ax[2].set_xlim(10,40)
ax[2].set_ylim(0,1.1)
ax[2].set_xticks([20,30])
ax[2].grid(False)

plt.tight_layout()
plt.show()

print("\nTemperatur 32.5°C adalah:")
print(round(mu_panas*100),"% Panas")
print(round(mu_sedang*100),"% Sedang")
print(round(mu_dingin*100),"% Dingin")
```
